```python
import numpy as np
import matplotlib.pyplot as plt
import pandas as pd
import warnings
warnings.filterwarnings("ignore")
```

```python
with open("utility_record_0.txt") as textFile:
    elessar_0 = [line.split(", ") for line in textFile]

with open("utility_record_0.25.txt") as textFile:
    elessar_0_25 = [line.split(", ") for line in textFile]
    
with open("utility_record_0.5.txt") as textFile:
    elessar_0_5 = [line.split(", ") for line in textFile]
    
with open("utility_record_0.75.txt") as textFile:
    elessar_0_75 = [line.split(", ") for line in textFile]
    
with open("utility_record_1.txt") as textFile:
    elessar_1 = [line.split(", ") for line in textFile]
    
with open("utility_record_primary.txt") as textFile:
    primary = [line.split(", ") for line in textFile]
    
with open("utility_record_conservative.txt") as textFile:
    conservative = [line.split(", ") for line in textFile]
```

```python
elessar_np_0 = np.array(elessar_0)
elessar_np_0_25 = np.array(elessar_0_25)
elessar_np_0_5 = np.array(elessar_0_5)
elessar_np_0_75 = np.array(elessar_0_75)
elessar_np_1 = np.array(elessar_1)

primary_np = np.array(primary)
conservative_np = np.array(conservative)
```

```python
elessar_np_0 = elessar_np_0.reshape(1000,-1)
elessar_np_0_25 = elessar_np_0_25.reshape(1000,-1)
elessar_np_0_5 = elessar_np_0_5.reshape(1000,-1)
elessar_np_0_75 = elessar_np_0_75.reshape(1000,-1)
elessar_np_1 = elessar_np_1.reshape(1000,-1)

primary_np = primary_np.reshape(1000,-1)
conservative_np = conservative_np.reshape(1000,-1)
```

```python
for i in range(1000):
    elessar_np_0[i][0] = elessar_np_0[i][0][1:]
    elessar_np_0[i][-1] = elessar_np_0[i][-1][:-1]
    
    elessar_np_0_25[i][0] = elessar_np_0_25[i][0][1:]
    elessar_np_0_25[i][-1] = elessar_np_0_25[i][-1][:-1]
    
    elessar_np_0_5[i][0] = elessar_np_0_5[i][0][1:]
    elessar_np_0_5[i][-1] = elessar_np_0_5[i][-1][:-1]
    
    elessar_np_0_75[i][0] = elessar_np_0_75[i][0][1:]
    elessar_np_0_75[i][-1] = elessar_np_0_75[i][-1][:-1]
    
    elessar_np_1[i][0] = elessar_np_1[i][0][1:]
    elessar_np_1[i][-1] = elessar_np_1[i][-1][:-1]
    
    primary_np[i][0] = primary_np[i][0][1:]
    primary_np[i][-1] = primary_np[i][-1][:-1]
    conservative_np[i][0] = conservative_np[i][0][1:]
    conservative_np[i][-1] = conservative_np[i][-1][:-1]
```

```python
elessar_np_0[0][0] = elessar_np_0[0][0][1:]
elessar_np_0[-1][-1] = elessar_np_0[-1][-1][:-2]

elessar_np_0_25[0][0] = elessar_np_0_25[0][0][1:]
elessar_np_0_25[-1][-1] = elessar_np_0_25[-1][-1][:-2]

elessar_np_0_5[0][0] = elessar_np_0_5[0][0][1:]
elessar_np_0_5[-1][-1] = elessar_np_0_5[-1][-1][:-2]

elessar_np_0_75[0][0] = elessar_np_0_75[0][0][1:]
elessar_np_0_75[-1][-1] = elessar_np_0_75[-1][-1][:-2]

elessar_np_1[0][0] = elessar_np_1[0][0][1:]
elessar_np_1[-1][-1] = elessar_np_1[-1][-1][:-2]

primary_np[0][0] = primary_np[0][0][1:]
primary_np[-1][-1] = primary_np[-1][-1][:-2]
conservative_np[0][0] = conservative_np[0][0][1:]
conservative_np[-1][-1] = conservative_np[-1][-1][:-2]
```

```python
elessar_float_0 = elessar_np_0.astype(np.float)
elessar_float_0_25 = elessar_np_0_25.astype(np.float)
elessar_float_0_5 = elessar_np_0_5.astype(np.float)
elessar_float_0_75 = elessar_np_0_75.astype(np.float)
elessar_float_1 = elessar_np_1.astype(np.float)

primary_float = primary_np.astype(np.float)
conservative_float = conservative_np.astype(np.float)
```

```python
elessar_mean_0 = np.mean(elessar_float_0, axis=0)
elessar_mean_0_25 = np.mean(elessar_float_0_25, axis=0)
elessar_mean_0_5 = np.mean(elessar_float_0_5, axis=0)
elessar_mean_0_75 = np.mean(elessar_float_0_75, axis=0)
elessar_mean_1 = np.mean(elessar_float_1, axis=0)

primary_mean = np.mean(primary_float, axis=0)
conservative_mean = np.mean(conservative_float, axis=0)
```
